```python
from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
import time, re

webdriverpath = "/home/eesha/Softwares/chromedriver_linux64/chromedriver"
url = "https://www.youtube.com/watch?v=eN6AYHAT8UM"
```

```python
driver = webdriver.Chrome(webdriverpath) 
driver.get(url)
driver.execute_script("window.scrollTo(0,document.documentElement.scrollHeight);")
```

```python
title = driver.find_elements_by_xpath('//*[@id="container"]/h1/yt-formatted-string')[0].text
print(title)

comments = driver.find_elements_by_xpath('//*[@id="count"]/yt-formatted-string')
num_comments = 0
for c in comments:
    num_comments = c.text
    num_comments = num_comments.split(" ")[0]
    num_comments = re.sub(r"[,]+", '', num_comments)
    num_comments = int(num_comments)
    print(num_comments)
```

```python
comments = []
likes_count = []
replies_count = []
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.documentElement.scrollHeight")
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0,document.documentElement.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
    comment = driver.find_elements_by_xpath('//*[@id="content-text"]')
    metadata = driver.find_elements_by_xpath('//*[@id="vote-count-middle"]')
    replies = driver.find_elements_by_xpath('//*[@id="replies"]/ytd-comment-replies-renderer')
    for c, md, rp in zip(comment, metadata, replies):
        comment_text = c.text.replace('\n', ' ')
        comment_likes = md.text
        comment_replies = rp.text.split(" ")[1].split(" ")[0]
        if comment_replies.isnumeric() == False:
            comment_replies = '0'
        print(comment_text, comment_likes, comment_replies)
        print()
        comments.append(comment_text)
        likes_count.append(comment_likes)
        replies_count.append(comment_replies)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
       print("break")
       break
    last_height = new_height

num_comments = len(comments)
print(num_comments)
```

```python
import csv
with open('youtube_comments_test.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Comment", "Likes", "Replies"])
    for comment, likes, replies in zip(comments, likes_count, replies_count):
        if 'K' in likes:
            likes = likes.lower().split('k')[0]
            likes = likes + '000'
        writer.writerow([comment, likes, replies])
```

```python
scraped_data = {}
scraped_data['title'] = title
scraped_data['number of comments'] = num_comments
all_comments = [{} for i in range(num_comments)]
id = 0
for comment, likes, replies in zip(comments, likes_count, replies_count):
    if 'K' in likes:
        likes = likes.lower().split('k')[0]
        likes = likes + '000'
    all_comments[id] = {'id': id, 'rt': 'None', 'comment': comment, 'number of likes': likes, 'number of replies': replies}
    id += 1
scraped_data['comments'] = all_comments
print(scraped_data)
```
